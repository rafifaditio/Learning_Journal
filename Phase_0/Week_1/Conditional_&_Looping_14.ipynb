```python
i = 1
j = 1

while i <= 10:
  print(f'perkalian {i}') 
  while j <= 10:
    print(f'{i} x {j} = {i * j}')
    j = j + 1
  print('-'*10)
  i = i + 1
  j = 1
```

```python
cities = cities + ['Bandung', 'Palembang', 'Balik Papan']
cities
```

### **Control Flow Statement**

**Continue**

```python
i = 0

while i <= 10:
  i = i + 1
  if i % 2 == 1:
    continue
  print(i)
```

**Break**

```python
i = 0

while i <= 10:
  i = i + 1
  print(i)
  if i == 7:
    break
```
